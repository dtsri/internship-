**Sample sales data cleaning**

```python
#using sample sales data
sales = pd.read_csv('/content/sales_data_sample.csv')
```

```python
#using encoding feature to read and view the sales data
sales = pd.read_csv('/content/sales_data_sample.csv', encoding='latin1')
```

```python
#to view the data of the first five rows
sales.head()
```

```python
#using tail to view the last records
sales.tail()
```

```python
#to know the shape
sales.shape
```

```python
#checking the datatypes
sales.dtypes
```

```python
#to check the information
sales.info()
```

```python
#so there are a lot of datatypes to be converted into their respective datatypes like phone to int, etc
#checking null and missing values
sales.isnull().sum()
```

```python
#dropping addressline 2 and state and territory columns as country and city columns are there and seem more important
#and addressline 2 , state and territory have too many null/empty values
sales.drop(['ADDRESSLINE2', 'STATE', 'TERRITORY'], axis=1, inplace=True)
```

```python
sales.isnull().sum()
```

```python
#dropping null values in postal code
sales.dropna(subset=['POSTALCODE'], inplace=True)
```

```python
#checking duplicates values
sales.duplicated().sum()
```

```python
#now there are no duplicates, null or missing values
#converting datatypes
```

```python
#converting the orderdate to date datatype
sales['ORDERDATE'] = pd.to_datetime(sales['ORDERDATE'])
```

```python
#converting phone to the number datatype
sales['PHONE'] = pd.to_numeric(sales['PHONE'])
```

```python
#since some numbers are starting with 0 and are international its not accepting so converting to string
sales['PHONE'] = sales['PHONE'].astype(str)
```

```python
#removing spaces or the dots in phone column
sales['PHONE'] = sales['PHONE'].str.replace('.', '')
```

```python
# Converting the cleaned phone column to numeric, coercing errors to NaN
sales['PHONE'] = pd.to_numeric(sales['PHONE'], errors='coerce')

# Check for any NaN values introduced by the conversion
print("NaN values in PHONE after numeric conversion:", sales['PHONE'].isnull().sum())
```

```python
# Dropping rows with NaN values only in the 'PHONE' column
sales.dropna(subset=['PHONE'], inplace=True)

# Check the number of null values in the 'PHONE' column after dropping
print("NaN values in PHONE after dropping:", sales['PHONE'].isnull().sum())

# Check the shape of the DataFrame to see how many rows were removed
print("Shape of DataFrame after dropping rows with NaN in PHONE:", sales.shape)
```
